- mean
> The mean is the average of all the observations in the data.
- median
> It is a middle obeservation in a list of observations sorted either in ascending or decending order.
- mode
> Mode is the observation or value that occurs the most number of times in a dataset.
- central tendency
> It is a single value that tries to describe the data by odentifying the central position within ghe dataset. Mean, median and mode are all measures of central tendency.
- distribution
> A distribution describes how values are distributed, which are more common and the ones that are uncommon.
- variance
> variance measures how the numbers in a dataset are spread out. It measures how far each number is from the mean.
- standard deviation
> The standard deviation is a measure of variance within a dataset. It is the square root of variance.

#### Fill in blanks with brief description of the following terms:
  
  
- **Correlation**
Correlation is a statistical measure that expresses the relationship between two variables or how each variable moves in relation to the other. It is measured by the correlation coefficient.
- **Correlation Coefficient**
It measures the correlation between two variables and can take values between absolute +1 and -1. If the correlation coefficient is +1, it means that for a given % change in variable 1; variable 2 also changes by the same %. When the correlation coefficient is a -1 it means that for a certain % increase in one variable, the value of the second variable falls by the same %.

### Step 2 - Review Hypothesis Testing & P-Value

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

#### Fill in blanks with brief description of the following terms:

- **Hypothesis**
> A hypothesis is a tentative statement that details the relationship between two or more variables

- **Null Hypothesis**
> The null hypothesis always states that there is no difference between the variables being tested and any observed difference is due to chance.

- **P-Value**
> P values measures the probability that the observed differences between two variables is due to chance.It helps in accepting or rejecting the null hypothesis. 

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
import pandas as pd
salary_df = pd.read_csv("salary_numeric.csv")
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

There is no significant relationship between education and salary

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(salary_df["salary"], salary_df["education"])
```

r = .477 and p-value = 0.0

#### Are these results statistically significant? How do you know?

Yes, there is a moderate positive correlation between salary and education indicating that higher the level of education; higher the salary. Since the p=value is <.05, the relationship is statistically significant.

### What is the correlation between salary and age?

There is no significant relationship between age and salary.

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(salary_df["salary"], salary_df["age"])
```

r = 0.66 and p-value = 0.0

#### Are these results statistically significant? How do you know? 

There is a strong positive correlation between salary and age which is statistically significant as the p-value is < 0.05. Higher the age; higher the salary of the manager.

### What is the correlation between salary and industry experience? 

There is no significant relationship between industry experience and salary.

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(salary_df["salary"], salary_df["industry_experience"])
```

r = 0.62 and p-value = 0.0

#### Are these results statistically significant? How do you know?

There is a  strong positive correlation between salary and industry experience because the statistic is > 0.50, which is statistically significant because the p=value is < 0.05. Higher the industry experience; higher the manager's salary.

### Step 4 - Report Findings

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### What was the null hypothesis for the relationships between education and salary, age and salary, and industry experience and salary?

Null Hypothesis:
1. There is no relationship between education and salary.
2. There is no relationship between age and salary.
3. There is no relationship between industry experience and salary.

### Can you reject the null hypothesis for any of the relationships you explored above?

Yes, we can reject all the 3 null hypothesis above and accept the alternative hypothesis that there is a relationship between education and salary, age and salary, and industry experience and salary.

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
import numpy as np
pd.crosstab(index=salary_df["industry"], columns= salary_df["education"], values=salary_df["salary"], aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
crosstab=pd.crosstab(index=salary_df["industry"], columns= salary_df["education"], values=salary_df["salary"], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level
plt.figure(figsize=(12,8))
ax = sns.heatmap(crosstab, annot=True, linewidths=.5, cmap='YlGnBu', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18);
plt.show()
```

##### Insights:

From the above heatmap we can conclude:
1. For all levels of education, Manager salaries are highest in the Tech industry and lowest in the healthcare industry.
2. For the same industry, manager salaries increase based on the education level of the manager. Managers with a Master's degree earn more than 2 times the salary of those with just a high school diploma.
